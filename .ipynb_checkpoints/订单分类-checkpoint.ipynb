```python
import os
import re
import time
import datetime
import numpy as np

import pandas as pd
```

```python
#订单所在目录
order_dir = r"D:\Downloads"


# 供应商路径
supplier_base_path = r"D:\奇货居\素材\商城图片素材\供应商详情.xlsx"

# 输出目录
out_dir = r"D:\奇货居\work\外发订单信息"


order_date_patt = '^订单(?P<date>20\d{2}(?:-\d{1,2}){2}\D+?\d{1,2}(?:_\d{1,2}){1,2})\.xls[xm]?$'
date_format = "%Y-%m-%d %H_%M_%S"

# 可入系统 目录

in_sys_dir = r'D:\奇货居\work\运单信息\可入系统'

```

```python


# 获取最新信息
def get_new_file_path(file_dir,file_date_patt,date_format,recursion=False):
    
    order_com = re.compile(file_date_patt)

    date = pd.to_datetime("1900-1-1")
    max_date_file_path = ""
    
    # 
    if recursion:
        # 递归遍历
        _file_dir = file_dir
        for file_dir,dir_name,file_names in os.walk(_file_dir):

            for file_name in file_names:
                order_date = order_com.match(file_name)
                if order_date:
                    order_date = order_date.groupdict()['date']

                    order_date = pd.to_datetime(order_date,format= date_format)
                    if order_date > date:
                        date = order_date
                        max_date_file_path = os.path.join(file_dir,file_name)
    else:
        # 遍历当前目录
        for file_name in os.listdir(file_dir):

            order_date = order_com.match(file_name)

            if order_date:
                order_date = order_date.groupdict()['date']

                order_date = pd.to_datetime(order_date,format= date_format)
                if order_date > date:
                    date = order_date
                    max_date_file_path = os.path.join(file_dir,file_name)

    

    return max_date_file_path,date


```

```python


new_order_path,new_order_date = get_new_file_path(order_dir,order_date_patt,date_format)
new_order_df = pd.read_excel(new_order_path)
new_order_df.replace({"订单号":{"\s":""}},regex=True,inplace=True)
new_order_df['供应商ID'] = new_order_df['商品ID'].str[:5]
```

```python
# 关联供应商信息表
gys_df = pd.read_excel(supplier_base_path).dropna()
new_order_r_df = pd.merge(new_order_df ,gys_df[['供应商ID','供应商']],how='left',on="供应商ID")
new_order_r_df['运单号'] = ""

# 消除 未来得及的运单信息
in_sys_patt = r"可入系统运单信息 (?P<date>\d{4}-\d{1,2}-\d{1,2} \d{1,2}_\d{1,2}_\d{1,2})\.xls[xm]?"
in_sys_path,new_insys_date = get_new_file_path(in_sys_dir,in_sys_patt,date_format)
in_sys_df = pd.read_excel(in_sys_path)

in_sys_df.replace({"订单号":{"\t":""}},regex=True,inplace=True)

new_order_r_df = new_order_r_df[new_order_r_df['订单号'].isin( set(new_order_r_df['订单号'])
                                                           - set(in_sys_df['订单号']) )]


# 存储外发订单
new_date_s = new_order_date.strftime("%Y-%m-%d %H_%M_%S")

order_total_path = os.path.join(out_dir,"总 外发订单 %s.xlsx"  %new_date_s)
new_order_r_df.to_excel(order_total_path,index=False) 

```

```python
# 供应商 分发文件
fields = [ '订单号',"运单号",'商品名', '下单用户', '下单时间', '规格',  '收件人电话','收件人地址','供应商']

data_r = new_order_r_df[fields]

for supplier in data_r['供应商'].unique():
    supplier_path = os.path.join(out_dir,"外发订单 %s %s.xlsx" %(supplier,new_date_s))
    temp = data_r[data_r['供应商'] == supplier]
    del temp['供应商']
    temp.to_excel(supplier_path,index=False)
   
```
